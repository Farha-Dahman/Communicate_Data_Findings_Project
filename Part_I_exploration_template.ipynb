### Question 2: How does trip duration vary by user type?

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x='user_type', y='duration_sec')
plt.title('Trip Duration by User Type', fontsize=12)
plt.xlabel('User Type', fontsize=11)
plt.ylabel('Duration (seconds)', fontsize=11)
plt.yscale('log')
plt.show()
```

> **Observations:**

> **Trip Duration Variation:**
> Customers generally take longer trips compared to Subscribers.
The median trip duration for Customers is higher than that for Subscribers.
The interquartile range (IQR) for Customers is also wider, indicating more variability in their trip durations.

> **Outliers:**
> Both user types exhibit some very long trips, as seen in the outliers extending beyond the whiskers of the box plots.
Subscribers have more prominent outliers compared to Customers, suggesting a subset of long-duration trips.

### Question 3: What are the correlations between numerical features?

```python
# Question: How does the number of trips vary by user type and gender?
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='user_type', hue='member_gender')
plt.title('User Type and Gender Distribution', fontsize=14)
plt.xlabel('User Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender')
plt.show()
```

**Observation:** Among both "Subscribers" and "Customers," males tend to have the highest number of trips especially from the "Subscriber" user type, followed by females and others.

## Multivariate Exploration

### Question 1: How does trip duration vary by gender and user type?

```python
# How does trip duration vary by gender and user type?
facet = sns.FacetGrid(data, col='member_gender', row='user_type', margin_titles=True)
facet.map(plt.hist, 'duration_sec', bins=20, color='purple')
facet.set_axis_labels('Duration (seconds)', 'Frequency')
facet.fig.subplots_adjust(top=0.9)
facet.fig.suptitle('Trip Duration by Gender and User Type', fontsize=11)
plt.show()
```

**Observation:** **Customers** (shown in the top row) tend to have longer trip durations compared to **Subscribers** (shown in the bottom row). This is evident from the distribution of trip durations, where the bars for Customers extend farther along the x-axis (duration) compared to those for Subscribers.

- **Gender Comparison:**
     - For all genders (Male, Female, Other), the trip durations of Customers show a wider spread compared to Subscribers, indicating that Customers generally take longer trips.
- **Subscribers vs. Customers:**
     - Subscribers tend to have more concentrated and shorter trip durations, possibly due to their routine usage for commuting or specific purposes.
- **Frequency:**
     - The frequency of Subscribers is much higher than that of Customers, as seen in the taller bars for Subscribers.

### Question 2: How does trip duration relate to user age and bike share program participation?

```python
# Question: How does trip duration relate to user age and bike share program participation?
# Calculate age from member_birth_year
data['age'] = 2025 - data['member_birth_year']
plt.figure(figsize=(9, 6))
sns.scatterplot(data=data, x='age', y='duration_sec', hue='bike_share_for_all_trip', alpha=0.6)
plt.title('Trip Duration by Age and Bike Share Participation', fontsize=12)
plt.xlabel('Age', fontsize=11)
plt.ylabel('Duration (seconds)', fontsize=11)
plt.legend(title='Bike Share Program')
plt.yscale('log')
plt.show()
```

**Observation:**
- Younger users (2040 years old) are more likely to participate in the bike share program.
- Older users tend to have more variability in trip durations, including both very short and long trips.
- Participants in the bike share program generally take shorter trips.

### Talk about some of the relationships you observed in this part of the investigation. Were there features that strengthened each other in terms of looking at your feature(s) of interest?

> The relationship between trip duration, user type, and gender revealed that these features interact in meaningful ways. Customers consistently demonstrated longer trip durations compared to Subscribers across all genders. This indicates that Customers may use the bike-sharing service for more recreational or leisurely activities, while Subscribers likely use it for more routine, time-constrained purposes like: commuting. Gender further highlighted differences within user types, with male and female Customers exhibiting similar trip duration patterns.

> When looking at trip duration, user age, and bike share program participation, a strong relationship emerged. Younger users (2040 years old) were more likely to participate in the bike share program and typically took shorter trips, likely using the service for commuting. This interaction strengthens the idea that the bike share program appeals to those seeking convenience and quick trips.

### Were there any interesting or surprising interactions between features?

> Gender and User Type with Trip Duration: It was surprising to see how concentrated and short trip durations were for Subscribers compared to Customers, regardless of gender. This pattern indicates that Subscribers, regardless of demographic differences, share similar usage patterns, likely tied to routine travel.
